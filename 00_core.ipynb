```python
n = 200
fibonacci = np.zeros((n))
fibonacci[0] = 0
fibonacci[1] = 1
index = 2
for index in range(2, n):
    fibonacci[index] = fibonacci[index-2] + fibonacci[index-1]
```

```python
class HelloSayer():
    "Say hello `to` using `sayHello`"
    def __init__(self, to): self.to = to
    def say(self): sayHello(self.to)
    
```

```python
show_doc(HelloSayer.say)
```

```python
o = HelloSayer("Alexis")
o.say()
```
